```python
#checking for missing values in my dataframe, focus on year 2014
fbs.isna().sum()
missing_Y2014 = fbs['Y2014'].isna().sum()
total_count = fbs['Y2014'].count() + missing_Y2014

#calculating the percentage of missing values
percentage_missing = (missing_Y2014 / total_count) * 100
round(percentage_missing, 3)
```
